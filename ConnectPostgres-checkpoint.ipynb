```python
#Zadanie C
quarters_count = pd.read_sql('WITH uslugi AS (SELECT EXTRACT(QUARTER FROM z.data_nadania) AS kwartal, u.nazwa AS usluga, COUNT(*) AS liczba_uslug FROM poczta_oltp.zakup z INNER JOIN poczta_oltp.usluga u USING(id_uslugi) WHERE EXTRACT(YEAR FROM z.data_nadania) =  2018 GROUP BY 1, 2) SELECT kwartal, liczba_uslug FROM uslugi u ORDER BY 1;',engine)
```

```python
quarters_count.columns
```

```python
quarters_count_R = quarters_count.groupby('kwartal').sum()[['liczba_uslug']]
```

```python
quarters_count_R['liczba_uslug'].keys()
quarters_count_R['liczba_uslug'].values
```

```python
#dodanie indexu łańcuchowego do

index = []





for i in range(len(quarters_count_R['liczba_uslug'].values)):
    if i == 0:
        index.append(1)
    else:
        index.append(round(quarters_count_R['liczba_uslug'][i+1]/quarters_count_R['liczba_uslug'][i],2))
    
        


quarters_count_R['index chain'] = index

        
```
